# Test Monitor Variable

```python
import numpy as np
import time
import matplotlib.pyplot as plt

import lightlab.util.characterize as chize
```

```python
# Turn this on when executing interactively
livePlots = False
```

```python
def osc(period):
    return np.sin(2 * np.pi * time.time() / period)

def variables():
    return [osc(2), osc(3), osc(4)]

if livePlots:
    sleepSec = .01
    plotEvery = 10
else:
    sleepSec = .001
    plotEvery = 100
chize.monitorVariable(variables, sleepSec=sleepSec, nReps=100, plotEvery=plotEvery)
```
